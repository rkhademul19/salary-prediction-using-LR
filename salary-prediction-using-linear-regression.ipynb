```python
# import library
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/kaggle/input/datasets/rkhademul19/salary-prediction/Salary Data.csv")
df.head()
```

```python
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title('Salary Prediction')
```

```python
x=df[["Experience Years"]]
```

```python
y=df[['Salary']]
```

```python
# Apply Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
plt.scatter(x.mean(), y.mean(), color='red')
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title("Salary Prediction")
```

```python
# training the model
reg.fit(x,y)
```

```python
reg.predict([[12.5]])
```

```python
new_df = pd.DataFrame({'Experience Years': [1.1]})
new_df['predict_salary'] = reg.predict(new_df)
new_df
```

```python
df['predict_salary']=reg.predict(x)
df.head()
```

```python
plt.plot(x,reg.predict(x))
plt.scatter(x.mean(), y.mean(), color="red")
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Slary Prediction")
```

```python
df['lose']=df['Salary']-df['predict_salary']
df.head()
```

```python
# Lost and Cost function
from sklearn.metrics import mean_squared_error, mean_absolute_error

```

```python
mse=mean_squared_error(df['Salary'], df['predict_salary'])
mse
```

```python
# Regression accuracy calculation

reg.score(x,y)
```

```python
plt.plot(x, reg.predict(x))
plt.scatter(x.mean(),y.mean(), color='green')
plt.scatter(df['Experience Years'], df['Salary'])
plt.scatter(df['Experience Years'], reg.predict(x))
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Slary Prediction")
```

```python
from sklearn.metrics import r2_score
r2_score(y, reg.predict(x))
```
